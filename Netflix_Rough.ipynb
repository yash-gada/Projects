# Data Cleaning and Processing

```python
#netflix["date_added"] = pd.to_datetime(netflix["date_added"])
#netflix.sort_values(by = "release_year")
#(pd.to_datetime(netflix["date_added"].fillna(method = "ffill")).dt.year < netflix["release_year"]).sum()
#netflix[netflix["date_added"].isnull() == False]["date_added"].dt.year

netflix["date_added"] = pd.to_datetime(netflix["date_added"])
```

```python
netflix = netflix.sort_values(by="release_year")
netflix.head()
```

```python
netflix["date_added"] = netflix["date_added"].fillna(method="bfill")
netflix["date_added"].isnull().sum()
```

```python
netflix["date_added_year"] = netflix["date_added"].dt.year
netflix["date_added_month"] = netflix["date_added"].dt.month
netflix["date_added_week"] = netflix["date_added"].dt.week
netflix.head()
```

```python
netflix["release_year"].isnull().sum()
```

```python
netflix = netflix.fillna(value = "NA")
```

```python
netflix.isnull().sum()
```

```python
netflix.info()
```

### *Unnesting data by exploding*




```python
netflix["cast"] = netflix["cast"].str.split(",")
netflix["listed_in"] = netflix["listed_in"].str.split(",")
netflix["director"] = netflix["director"].str.split(",")
netflix["country"] = netflix["country"].str.split(",")
# cast, listed_in, director, country have multiple values in single cell
#8807 rows currently
```

```python
netflix.head()
```

```python
netflix = netflix.explode(["cast"], ignore_index=True)
netflix = netflix.explode(["listed_in"], ignore_index=True)
netflix = netflix.explode(["director"], ignore_index=True)
netflix = netflix.explode(["country"], ignore_index=True)

# exploding all the multiple value columns
```

```python
netflix.head()
```

```python
netflix.drop_duplicates(inplace = True)
```

```python
netflix.shape
```

```python
netflix.isnull().sum()
```

### Change the duration column to integer for easy analysis



```python
# Separating each cell of duration column by space
# Taking only the first index (integer part) and exclusing the string (min/season)

netflix["duration"] = netflix["duration"].str.split().str[0]
netflix
```

```python
# Trimming strings in each cell to remove any trailing or leading spaces

#netflix.select_dtypes(include="object").apply(lambda x:x.str.strip())
str_df = netflix.select_dtypes(include="object")
netflix[str_df.columns] = netflix.select_dtypes(include="object").apply(lambda x:x.str.strip())
del str_df
```

```python
# dropping duplicates before separating dataset

netflix.drop_duplicates(inplace = True)
netflix.count()
# 202010
```

```python
# Count of Movies for confirmation before separating the dataset

netflix[netflix["type"] == "Movie"].count()
```

```python
# Count of TV Shows for confirmation before separating the dataset

netflix[netflix["type"] == "TV Show"].count()
```
